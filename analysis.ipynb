```python
required_data = data.drop(columns=['id', 'symbol','volume','open_interest', 'exchange'])
```

```python
required_data.head()
```

```python
required_data['timestamp'] = pd.to_datetime(required_data['timestamp'])

required_data['year'] = required_data['timestamp'].dt.year
required_data['month'] = required_data['timestamp'].dt.month
required_data['day'] = required_data['timestamp'].dt.day
required_data['hour'] = required_data['timestamp'].dt.hour
required_data['minute'] = required_data['timestamp'].dt.minute

required_data.head()
```

```python
for i in ['year', 'month', 'day', 'hour', 'minute']:
    print(required_data[i].value_counts())
    
```
